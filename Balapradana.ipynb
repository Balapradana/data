```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data
file_path = "/content/btc_data.csv"
df = pd.read_csv(file_path)

# Data Cleaning
df = df.iloc[2:].copy()
df.columns = ["Date", "Close", "High", "Low", "Open", "Volume"]
df["Date"] = pd.to_datetime(df["Date"])
numeric_cols = ["Close", "High", "Low", "Open", "Volume"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Set Style
sns.set_style("whitegrid")
plt.figure(figsize=(15, 10))

# 1. Tren Harga Bitcoin
plt.subplot(2, 3, 1)
plt.plot(df["Date"], df["Close"], label="Close Price", color="blue")
plt.xlabel("Date")
plt.ylabel("Close Price ($)")
plt.title("Tren Harga Bitcoin")
plt.legend()

# 2. Histogram Harga Penutupan
plt.subplot(2, 3, 2)
sns.histplot(df["Close"], bins=30, kde=True, color="green")
plt.xlabel("Close Price ($)")
plt.ylabel("Frequency")
plt.title("Distribusi Harga Penutupan Bitcoin")

# 3. Boxplot Harga Harian
plt.subplot(2, 3, 3)
sns.boxplot(data=df[["Open", "High", "Low", "Close"]])
plt.title("Boxplot Harga Harian Bitcoin")

# 4. Scatter Plot Volume vs Harga Penutupan
plt.subplot(2, 3, 4)
sns.scatterplot(x=df["Volume"], y=df["Close"], alpha=0.5, color="red")
plt.xlabel("Volume")
plt.ylabel("Close Price ($)")
plt.title("Volume vs Harga Penutupan")
plt.xscale("log")

# 5. Moving Average 30 Hari
plt.subplot(2, 3, 5)
df["Close_MA30"] = df["Close"].rolling(window=30).mean()
plt.plot(df["Date"], df["Close"], label="Close Price", color="gray", alpha=0.5)
plt.plot(df["Date"], df["Close_MA30"], label="30-day MA", color="orange")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.title("Moving Average 30 Hari Bitcoin")
plt.legend()

# Menyesuaikan layout
plt.tight_layout()
plt.show()
```
